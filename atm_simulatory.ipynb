```python
actual_pin=576213
balance=50000
tries=0
print("Welcome to ATM Simulator\n",
    "Place your card")
while tries<3:
    pin=int(input("Enter your PIN"))
    if actual_pin==pin:
        while True:
            print("1.Check Balance\n","2.Withdraw\n","3.Deposit\n","4.Exit\n","(1,2,3,4)")
            choice=int(input("Enter your need"))
            if choice==1:
                print("The amount available is ",balance)
            elif choice==2:
                withdraw=int(input("Enter the withdraw amount"))
                if withdraw<=balance:
                    balance=balance-withdraw
                    print("The amount available is",balance)
                else:
                    print("Insufficient amount")
            elif choice==3:
                deposit=int(input("Enter the amount to  deposit"))
                balance=balance+deposit
                print("The amount available is:",balance)
            elif choice==4:
                print("Thank you for using ATM")
                break
            else:
                print("Invalid Input")
        break
    else:
        tries+=1 
        print("Invalid password,remaining attempts",3-tries)
else:
    print("Card Blocked for multiple failed attempts")
```
